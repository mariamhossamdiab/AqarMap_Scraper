```python
import requests 
from bs4 import BeautifulSoup
import csv
import pandas as pd
import openpyxl
res=[]
t=1
```

```python
a = input ("please enter {'r'} for-rent & {'s'} for-sale")
if (a=='r'):
    a='for-rent'
elif (a=='s'): 
    a='for-sale'
b=input ("please enter aqar type ( apartment ,furnished-apartment ,chalet ,villa ,land-or-farm ,building , commercial, administrative, medical, shared-rooms, land-or-commercial, property-type ) ")
c=input("enter location ")
p='?page='
d='1'
website=(f"https://aqarmap.com.eg/en/{a}/{b}/{c}/{p+d}")
page=requests.get(website)
x=website.split('/')

sale_or_rent=x[4]
Type=x[5]
print(website + Type)
```

```python
def main(page):
    src=page.content
    soup=BeautifulSoup(src,"lxml")
    aqars=soup.find_all('div',{'class':'search-listing-card__wrapper'})
    for aqar in aqars :
        link=aqar.find('a', {'class':'search-listing-card__container__link'} ).get('href')
        aqarprice=aqar.find('div',{'class':'search-listing-card__price'}).text.strip()
        Adtitle=aqar.find('h2',{'class':'search-listing-card__title'}).text.strip()
        aqaraddress=aqar.find('p',{'class':'search-listing-card__address'}).text.strip()
        aqar_=aqar.find_all('label',{'class':'search-listing-card__attributes_item'})
        #print(len(aqar_))
        #print(aqar_)
        aqararea='nan'
        aqarbedrooms='nan'
        aqarbathrooms='nan'
        a=['nan','nan','nan']
        for i in range(0,len(aqar_)):
            a[i]=aqar_[i].text.strip()
            #print(a[i])
        aqararea=str(a[0])
        aqarbedrooms=str(a[1])
        aqarbathrooms=str(a[2])
        if (Type == 'building' or Type == 'land-or-farm'or Type =='apartment'or Type == 'villa' or Type == 'chalet') : 
            aqar_type='residential'
        elif (Type=='townhouse' or Type=='furnished-apartment'or Type=='shared-rooms' ):
            aqar_type='residential'
        elif (Type=='land-or-commercial'or Type=='commercial'):
            aqar_type='commercial'
        elif (Type=='administrative'):
            aqar_type='administrative'
        elif (Type=='medical' ):
            aqar_type='medical'
        else :
            aqar_type='property-type'
        print(link)
        res.append ({"link":('https://aqarmap.com.eg/'+link),"Adtitle":Adtitle,"price":aqarprice,"address":aqaraddress  ,"area":aqararea,'bedrooms':aqarbedrooms,'bathrooms':aqarbathrooms,'sale_or_rent':sale_or_rent,'unit':Type,'aqar type':aqar_type,'price/meter':'nan'})
for i in range(1,20):
    v=str(i)
    website=(f"https://aqarmap.com.eg/en/{a}/{b}/{c}/{p+v}")
    page=requests.get(website)
    main(page)    
```

```python
saveName='scraped.xlsx'

df = pd.DataFrame(res)
writer = pd.ExcelWriter(saveName)
df.to_excel(writer, sheet_name='Sheet1',index=False)
writer.save()
```
